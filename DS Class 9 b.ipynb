
# Scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, '-ok', color="Black")
```

```python
# Another way
plt.scatter(x,y)
```

# Histogram

```python
# histtype = bar, barstacked, step, stepfilled 
# default = bar
# bins tells the width (Zoom-in or Zoom-out)
data = np.random.randn(1000)
plt.hist(data, bins=10, histtype="stepfilled")
```

```python
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x), label = "Sine Graph")
plt.plot(x, np.cos(x), label = "Cosine Graph")

plt.legend(loc = "lower left", ncol = 2, fancybox = True, borderpad = 1, shadow = True)
```
