```python
def cast_len(val):
    if val == "None":
        return 0
    else:
        return len(str(val).split(", "))
```

```python
data["cast"].isnull().sum()
```

```python
data["cast"].replace(np.nan, "None",inplace=True)
```

```python
data.tail()
```

```python
data["cast_length"] = data["cast"].apply(cast_len)
```

# Matplotlib and Numpy

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
## To save this plot:
x = np.linspace(0,10,1000)

fig = plt.figure()

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

fig.savefig("LinePlot.png")

plt.show()
```

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
## To customize color:
y = np.linspace(0,10,1000)

fig = plt.figure()

plt.plot(y, np.sin(x), color = "blue")
plt.plot(y, np.sin(x-1), color = "red")
plt.plot(y, np.sin(x-2), color = "green")
plt.plot(y, np.cos(x), color = "black")

fig.savefig("Color_LinePlot.png")
```

```python
## To add linestyles
z = np.linspace(0,10,1000)

plt.plot(z, np.sin(x), color = "blue", linestyle = "solid")
plt.plot(z, np.sin(x-1), color = "red", linestyle = "dashed")
plt.plot(z, np.sin(x-2), color = "green", linestyle = "dotted")
plt.plot(z, np.cos(x), color = "black", linestyle = "dashdot")

```

```python
## Labeling of a graph
a = np.linspace(0,10,1000)

plt.plot(a, np.sin(x), color = "blue", linestyle = "solid")
plt.plot(a, np.sin(x-1), color = "red", linestyle = "dashed")
plt.plot(a, np.sin(x-2), color = "green", linestyle = "dotted")
plt.plot(a, np.sin(x-3), color = "black", linestyle = "dashdot")

plt.title("Sin Graph")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
a = np.linspace(0,10,1000)

plt.plot(a, np.sin(x), color = "blue", linestyle = "solid", label = "Sin(x)")
plt.plot(a, np.sin(x-1), color = "red", linestyle = "dashed", label = "Sin(x-1)")
plt.plot(a, np.sin(x-2), color = "green", linestyle = "dotted", label = "Sin(x-2)")
plt.plot(a, np.sin(x-3), color = "black", linestyle = "dashdot", label = "Sin(x-3)")

plt.title("Sin Graph")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
plt.show()
```
