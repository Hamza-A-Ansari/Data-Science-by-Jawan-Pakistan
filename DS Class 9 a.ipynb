# Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y, 'o', color = "black")
```

```python
## 2nd method
x = np.linspace(0,10,30)
y = np.sin(x)

plt.scatter(x,y)
```

```python
# scatter and line
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y, '-ok' , color = "black")
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
sizes = 1000 * np.random.randn(100)

plt.scatter(x, y, c=colors, s=sizes, alpha = 0.3)

plt.colorbar()
```

# Histogram

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
data = np.random.randn(1000)
plt.hist(data, bins = 30,alpha = 0.5, histtype = "stepfilled", color = 'steelblue')
```

```python
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x), '-b', label = "Sine Graph")
plt.plot(x, np.cos(x), '--r', label = "Cosine Graph")

plt.legend()
```

```python
## To change the position of the legend: (best, upper/lower left, upper/lower right, right, center, center left/right,
## lower/upper center)
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x), '-b', label = "Sine Graph")
plt.plot(x, np.cos(x), '--r', label = "Cosine Graph")

plt.legend(loc = "best")
```

```python
## If we waant to write legend in columns:
plt.plot(x, np.sin(x), '-b', label = "Sine Graph")
plt.plot(x, np.cos(x), '--r', label = "Cosine Graph")

plt.legend(loc = "best", ncol = 2)
```

```python
## 2D Plot
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],10000).T

plt.hist2d(x,y,bins = 20)
plt.show()
```

```python
## To add colours
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],10000).T

plt.hist2d(x,y,bins = 20, cmap = "Blues")
plt.show()
```
